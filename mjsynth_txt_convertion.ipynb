```python
nums_to_words = dict()

print("nums_to_words created")

with open('lexicon.txt', 'r') as lexicon:
    print("lexicon opened successfully")
    
    words_list = lexicon.readlines()
    print("words_list created")
    
    num_of_words = len(words_list)
    print("number of words in lexicon is", num_of_words)
    
    for i in range(num_of_words):
        nums_to_words[i] = words_list[i]
        
    print("the dictionary is filled")
    
    

path = ['annotation.txt', '/Users/bukharaevalimniazovich/Desktop/mjsynth_new/annotation_test.txt', '/Users/bukharaevalimniazovich/Desktop/mjsynth_new/annotation_train.txt', '/Users/bukharaevalimniazovich/Desktop/mjsynth_new/annotation_val.txt']        
    
print("-----------------")    
    
    
for p in range(1):    
    
    print("p =", p)
    
    # Read in the file
    with open(path[p], 'r') as file :
        lines = file.readlines()
        num_of_lines = len(lines)
    print("file was opened and read successfully")

    new_lines = []
    count = 1  
    
    for line in lines:
        line_split = line.split()
        new_line = line_split[0] + ' ' + nums_to_words[int(line_split[1])]
        new_lines.append(new_line)
        # Replace the target string
        if (count % 10000 == 0):
            print(count, "of", num_of_lines,"strings are replaced")
        count += 1
    
    # Write the file out again
    edge = 8919273 // 10
    with open(path[p], 'w') as f:
        for line in new_lines[0, edge]:
            f.write("%s\n" % line)
    
    
```
